**The goal of this notebook is to take the initial data from 'https://www.kaggle.com/datasets/saisaran2/rainfall-data-from-1901-to-2017-for-india' on rainfall data from India and Filter, Sort, and Visualize the data.** 

**First, I'll import the data and transform it into a Pandas Dataframe**

```python
import pandas as pd

def display(file):
    return(file)

InitialData = pd.read_csv("Rainfall_Data_LL.csv")
display(InitialData)
```

**Next, cleaning it up a bit**

**Let's change the dataframe so that the years are rows and the regions are coloumns**

```python
ReshapedData = InitialData.pivot_table(
        index='YEAR',        #sets years as rows
        columns='SUBDIVISION',  #sets regions as columns
        values='ANNUAL',     #values to populate the table
    )
display(ReshapedData)
```

**Now, let's try to visualize this data**

```python
import matplotlib.pyplot as plt

# Plot all regions
plt.figure(figsize=(14, 8))
for column in ReshapedData.columns:
    plt.plot(ReshapedData.index, ReshapedData[column], label=column)

plt.title('Annual Rainfall Trends (1901–2017)')
plt.xlabel('Year')
plt.ylabel('Annual Rainfall (mm)')
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))
plt.grid(True)
plt.show()
```

**Bar Chart**

```python
selected_years = ReshapedData.loc[ReshapedData.index % 40 == 0]

#plots the grouped bar chart
ax = selected_years.plot(kind='bar', figsize=(14, 8), width=0.8)

#sets plot labels and title
plt.xlabel('Year')
plt.ylabel('Annual Rainfall (mm)')
plt.title('Annual Rainfall Comparison Across Regions')

#adjusts legend
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')

#displays the plot
plt.tight_layout()
plt.show()
```

```python
selected_years = ReshapedData.loc[ReshapedData.index % 40 == 0] #change this to change the spacing of years
ax = selected_years.plot(kind='bar', stacked=True, figsize=(14, 8), width=0.8)

#set plot labels and title
plt.xlabel('Year')
plt.ylabel('Total Annual Rainfall (mm)')
plt.title('Total Annual Rainfall Stacked by Region')

#adjusts legend
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')

#displays the plot
plt.tight_layout()
plt.show()
```

**BOX PLOT**

```python
#creates a boxplot for each region
plt.figure(figsize=(12, 6))
ReshapedData.plot(kind='box', vert=True, figsize=(12, 6))

#sets plot labels and title
plt.title('Annual Rainfall Distribution by Region (1901–2017)')
plt.xlabel('Region')
plt.ylabel('Annual Rainfall (mm)')

#rotates x-axis labels for better readability
plt.xticks(rotation=90)

#displays the plot
plt.grid(True)
plt.show()
```

```python
#specific regions to analyze
regions_of_interest = ['Arunachal Pradesh', 'Naga Mani Mizo Tripura', 'Tamil Nadu']  #replace with chosen regions

#filters the DF for the selected regions
filtered_data = ReshapedData[regions_of_interest]

#creates boxplots for the selected regions
plt.figure(figsize=(10, 6))
filtered_data.plot(kind='box', vert=True, figsize=(10, 6))

#sets plot labels and title
plt.title('Annual Rainfall Distribution for Selected Regions (1901–2017)')
plt.xlabel('Region')
plt.ylabel('Annual Rainfall (mm)')

#displays the plot
plt.grid(True)
plt.show()
```
